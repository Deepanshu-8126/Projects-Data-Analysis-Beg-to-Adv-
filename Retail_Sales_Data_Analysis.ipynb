```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


data = {
    "Date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-02-01"],
    "Product": ["Mobile", "Laptop", "Tablet", "Mobile"],
    "Region": ["North", "South", "East", "North"],
    "Sales": [20000, 45000, 15000, 22000]
}

df=pd.DataFrame(data)
df
```

```python
# Total sales
total_sales = df['Sales'].sum()
print("Total Sales ",total_sales)
```

```python
df.groupby('Product')['Sales'].sum()
```

```python
df.groupby('Product')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Product ')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
# data cleaning
df.head()
df.info()
```

```python
df.loc[1, "Sales"] = None
df.loc[2, "Product"] = "mobile"
df
```

```python
#Missing values
df.isnull()
df.notnull()
```

```python
#Missing values count
print(df)
df.isnull().sum()
```

```python
#fill missing values
df['Sales']=df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df['Product']=df["Product"].str.capitalize()
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df)


#month , year
```

```python
#Month
df['Month'] = df['Date'].dt.month
df
```

```python
#Month wise sales
df.groupby("Month")["Sales"].sum()

# january total sales 54000
#february - 22000
```

```python

#Visualization
print(df)
a=df.groupby("Product")["Sales"].sum()
a
```
